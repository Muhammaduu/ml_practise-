# simple linear regresion

```python
#importing neccassary libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
```

```python
#loading the data
df = pd.read_csv("income.data.csv",index_col=0)
#head of the dataset
df.head()
```

```python
#tail of the dataset
df.tail()
```

```python
#cheking missing values
df.isnull().sum()
```

```python
# Independent variable (feature) and dependent variable (target)
X = df[['income']]
y = df['happiness']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
# Plotting the results
plt.scatter(X, y, color='blue') # Original data points
plt.plot(X_test, y_pred, color='red', linewidth=2) # Regression line
plt.xlabel('income (KSH)')
plt.ylabel('happiness')
plt.title('Linear Regression: income vs happiness')
plt.show()
```

# Logistic Regression

```python
# neccessary libraries 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
# loading data
df=pd.read_csv("purchase.csv")
df.head()
```

```python
df.tail()
```

```python
# Independent variable (feature) and dependent variable (target)
X = df[['annual_income']]
y = df['purchase']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating and training the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]
```

```python
# Evaluating the model
conf_matrix = confusion_matrix(y_test, y_pred)
6
class_report = classification_report(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")
print(f"ROC-AUC: {roc_auc}")
```

```python
# Plotting the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('annual income')
plt.ylabel('purchase')
plt.title('annual income vs purchase')
plt.legend(loc="lower right")
plt.show()
```

# Decision Tree

```python
# importing libraries
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# loading the dataset
df=pd.read_csv("decision_trees.csv")
df.head()
```

```python
# Convert categorical features to numeric
df['location'] = df['location'].map({'Nairobi': 1, 'Thika': 2, 'Lamu': 3,'Mombasa':4})
df['bedroom'] = df['bedroom'].map({'one bedroom': 1, 'two bed room': 2,'three bedroom':3})
```

```python
# Independent variables (features) and dependent variable (target)
X = df[["location","bedroom"]]
y = df['price']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating and training the decision tree model
model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
9
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")
```

```python
# Plotting the decision tree
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=['location', 'bedroom'])
plt.title('Decision Tree')
plt.show()
```

# Random Forest

```python
# neccessary libraries
from sklearn.ensemble import RandomForestClassifier
```

```python
# creating dic.
data = {
 'Age': [29, 45, 50, 39, 48, 50, 55, 60, 62, 43],
 'Cholesterol': [220, 250, 230, 180, 240, 290, 310, 275, 300, 280],
 'Max_Heart_Rate': [180, 165, 170, 190, 155, 160, 150, 140, 130, 148],
 'Heart_Disease': [0, 1, 1, 0, 1, 1, 1, 1, 1, 0]
}
# creatinf df
df = pd.DataFrame(data)
df.head()
```

```python
# Independent variables (features) and dependent variable (target)
X = df[['Age', 'Cholesterol', 'Max_Heart_Rate']]
y = df['Heart_Disease']
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Creating and training the random forest model
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")
```

```python
# Feature importance
feature_importances = pd.DataFrame(model.feature_importances_, index=X.columns,
columns=['Importance']).sort_values('Importance', ascending=False)
print(f"Feature Importances:\n{feature_importances}")
```

```python
# Plotting the feature importances
sns.barplot(x=feature_importances.index, y=feature_importances['Importance'])
plt.title('Feature Importances')
plt.xlabel('Feature')
12
plt.ylabel('Importance')
plt.show()
```
